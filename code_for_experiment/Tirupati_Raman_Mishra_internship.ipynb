```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
dat = pd.read_csv("tot.csv")
print(dat.shape)
dat.head()
```

```python
int_cat={'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'H':8}
dat["category"]=dat["category"].map(int_cat)

dat=dat[dat.views!="F"]
dat=dat[dat.likes!="F"]
dat=dat[dat.dislikes!="F"]
dat=dat[dat.comment!="F"]

dat["views"]=pd.to_numeric(dat["views"])
dat["likes"]=pd.to_numeric(dat["likes"])
dat["dislikes"]=pd.to_numeric(dat["dislikes"])
dat["comment"]=pd.to_numeric(dat["comment"])
print(dat.shape)
dat.dtypes
```

```python
#also u can use .isnull().any() 
print(dat.isna().any())
dat["published"]=LabelEncoder().fit_transform(dat["published"])
print(dat.dtypes)
dat.head()
```

```python
import time
import datetime

def checki(x):
    y = x[2:]
    h = ''
    m = ''
    s = ''
    mm = ''
    P = ['H','M','S']
    for i in y:
        if i not in P:
            mm+=i
        else:
            if(i=="H"):
                h = mm
                mm = ''
            elif(i == "M"):
                m = mm
                mm = ''
            else:
                s = mm
                mm = ''
    if(h==''):
        h = '00'
    if(m == ''):
        m = '00'
    if(s==''):
        s='00'
    bp = h+':'+m+':'+s
    return bp

def func_sec(time_string):
    h, m, s = time_string.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)


time = (dat["duration"].apply(checki)).apply(func_sec)
dat["duration"]=time
dat.head()
```
